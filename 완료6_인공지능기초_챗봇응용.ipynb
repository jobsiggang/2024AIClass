```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise  import cosine_similarity
sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

```

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/aidata/감성대화말뭉치.csv")
col=df['상황키워드']
col=col.unique()
print(col)
df=df[df["상황키워드"]=="진로,취업,직장"]
print(df)
```

```python
df["embedding"]=df["사람문장1"].map(lambda x:list(model.encode(x)))
print(df.head)
```

```python
!pip install gradio
```

```python
import gradio as gr
print("상황 선택이유 한문장, 학번 이름")

def chatbot (user):
  text_embedding=model.encode(user)
  df['similarity']=df['embedding'].map(lambda x:cosine_similarity([text_embedding],[x]).squeeze())
  answer=df.loc[df['similarity'].idxmax()]
  return answer["시스템문장1"]


demo=gr.Interface(fn=chatbot,inputs='text',outputs='text')
demo.launch(debug=True,share=True)
```
